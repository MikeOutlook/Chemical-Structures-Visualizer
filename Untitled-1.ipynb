```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display, HTML
import base64
from io import BytesIO


# 读取CSV文件
df = pd.read_csv('E:/11_AI_drug/test.csv')

# 创建HTML表格的头部
html_content = """
<style>
table {
    border-collapse: collapse;
    width: 100%;
}
th, td {
    border: 1px solid #ddd;
    padding: 8px;
    text-align: left;
}
th {
    background-color: #f2f2f2;
}
img {
    max-width: 200px;
    max-height: 150px;
}
</style>
<table>
<tr>
    <th>序号</th>
    <th>SMILES</th>
    <th>化学结构式</th>
</tr>
"""

# 处理每一行SMILES
for idx, smiles in enumerate(df['SMILES'], 1):
    try:
        # 将SMILES转换为分子对象
        mol = Chem.MolFromSmiles(smiles)
        
        if mol:
            # 绘制分子结构
            img = Draw.MolToImage(mol, size=(200, 150))
            
            # 将图像转换为base64编码
            buffered = BytesIO()
            img.save(buffered, format="PNG")
            img_str = base64.b64encode(buffered.getvalue()).decode()
            
            # 添加行到HTML表格
            html_content += f"""
<tr>
    <td>{idx}</td>
    <td style="font-family: monospace; font-size: 12px;">{smiles}</td>
    <td><img src="data:image/png;base64,{img_str}" alt="结构式"></td>
</tr>
"""
        else:
            html_content += f"""
<tr>
    <td>{idx}</td>
    <td style="font-family: monospace; font-size: 12px;">{smiles}</td>
    <td style="color: red;">无法解析的SMILES</td>
</tr>
"""
    except Exception as e:
        html_content += f"""
<tr>
    <td>{idx}</td>
    <td style="font-family: monospace; font-size: 12px;">{smiles}</td>
    <td style="color: red;">错误: {str(e)}</td>
</tr>
"""

html_content += "</table>"

# 显示HTML表格
display(HTML(html_content))

# 可选：将结果保存到新的CSV文件（包含图像路径）
print(f"已处理 {len(df)} 个分子结构")
```
